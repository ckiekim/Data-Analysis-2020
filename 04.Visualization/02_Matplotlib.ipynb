### 01. Matplotlib 기본 사용

```python
plt.plot([1, 3, 2, 4])
plt.ylabel('y-label')
plt.show()
```

#### 스타일 지정

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'bs:')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')
plt.show()
```

```python
t = np.arange(0., 5., 0.02)
plt.plot(t, t, 'r-', t, t**2, 'b-', t, t**3, 'g-')
plt.show()
```

#### 08. Matplotlib 그래프 영역 채우기

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_between(x[1:3], y[1:3], alpha=0.5)   # fill_between() 사용

plt.axhline(0); plt.axvline(0)
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_betweenx(y[2:4], x[2:4], color='pink', alpha=0.5)
plt.axhline(0); plt.axvline(0)
plt.show()
```

```python
x = np.arange(1.0, 4.0, 0.01)
y1 = x ** 2
y2 = x ** 2 / 2

plt.plot(x, y1); plt.plot(x, y2)
plt.xlabel('X-Label'); plt.ylabel('Y-Label')
plt.title('두 그래프 사이 영역 채우기')
plt.fill_between(x[100:201], y1[100:201], y2[100:201], color='lightgray', alpha=0.5)

plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 4, 8]

plt.plot(x, y1); plt.plot(x, y2)
plt.xlabel('X-Label'); plt.ylabel('Y-Label')
plt.title('임의의 영역 채우기')
plt.fill([1.9, 1.9, 2.5, 3.1, 3.1], [2, 5, 9, 11, 8], color='lightgray', alpha=0.5)

plt.show()
```

#### 09. Matplotlib 여러 곡선 그리기

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'r--', a, a**2, 'bo', a, a**3, 'g-.')
plt.show()
```

```python
plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.title('스타일 지정하기')
plt.show()
```

#### 10. Matplotlib 그리드 설정하기

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True)

plt.show()
```

```python
plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='y')
plt.title('축 지정하기')
plt.show()
```

```python
plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='y', color='red', alpha=0.5, linestyle='--')
plt.title('그리드 스타일 설정하기')
plt.show()
```

#### 11. Matplotlib 눈금 표시하기

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.xticks([0, 1, 2])
plt.yticks(np.arange(1, 6))

plt.show()
```

```python
plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2.1, 0.2), \
    labels=['Jan', '', 'Feb', '', 'Mar', '', 'April', '', 'May', '', 'June'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.show()
```

#### 12. Matplotlib 타이틀 설정하기

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='y', color='gray', alpha=0.5, linestyle='--')
plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)
plt.title('Sample graph', loc='right', pad=10)

plt.show()
```

```python
plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='y', color='gray', alpha=0.5, linestyle='--')
plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)
plt.title('Sample graph', loc='right', pad=20)
plt.title('Sample graph', fontdict=title_font, loc='left', pad=20)
plt.show()
```

#### 13. Matplotlib 수직선/수평선 표시하기

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.axhline(1, 0, 0.55, color='gray', linestyle='--', linewidth='1')
plt.axvline(1, 0, 0.16, color='lightgray', linestyle=':', linewidth='2')

plt.axhline(5.83, 0, 0.95, color='gray', linestyle='--', linewidth='1')
plt.axvline(1.8, 0, 0.95, color='lightgray', linestyle=':', linewidth='2')

plt.show()
```

```python
plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.hlines(4, 1, 1.6, colors='pink', linewidth=3)
plt.vlines(1, 1, 4, colors='pink', linewidth=3)

plt.show()
```

```python
X = np.linspace(0, 2*np.pi, 256)
S = np.sin(X)
C = np.cos(X)
plt.title("Sine, Cosine 그래프")
plt.plot(X, S, label="sine")
plt.plot(X, C, label="cosine")
plt.grid(True)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])  # LaTeX
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.legend(loc='best')  # loc = 0
plt.show()
```

#### 14. Matplotlib 막대 그래프 그리기

```python
x = np.arange(3)
years = ['2017', '2018', '2019']
values = [100, 400, 900]

plt.bar(x, values)
plt.xticks(x, years)
plt.show()
```

```python
plt.bar(x, values, width=0.6, align='edge', color="springgreen",
        edgecolor="gray", linewidth=3, tick_label=years, log=True)
plt.title('스타일 꾸미기')
plt.show()
```

#### 15. Matplotlib 수평 막대 그래프 그리기

```python
y = np.arange(3)
years = ['2017', '2018', '2019']
values = [100, 400, 900]

plt.barh(y, values, height=0.6, align='edge', color="springgreen",
        edgecolor="gray", linewidth=3, tick_label=years, log=False)
plt.show()
```

#### 16. Matplotlib 산점도 그리기

```python
np.random.seed(2020)    # Reproducible random state

N = 50
x = np.random.rand(N)
y = np.random.rand(N)

plt.scatter(x, y)
plt.show()
```

```python
np.random.seed(2020)    # Reproducible random state

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
import seaborn as sns 
iris = sns.load_dataset('iris')
x = iris[iris.species == 'setosa'].sepal_length
y = iris[iris.species == 'setosa'].sepal_width
plt.scatter(x, y)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Iris Setosa 품종')
plt.show()
```

#### 18. Matplotlib 히스토그램 그리기

```python
weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71, 80, 59, 67, 81, 69, 73, 69, 74, 70, 65]

plt.hist(weight)
plt.show()
```

```python
np.random.seed(2020)
x = np.random.randn(1000)
plt.hist(x)
plt.show()
```

```python
np.random.seed(2020)
x = np.random.randn(100000)
plt.hist(x, bins=100)
plt.show()
```

```python
x = np.linspace(-4.0, 4.0, 801)
nd = np.exp(-x*x/2) / np.sqrt(2*np.pi)
plt.plot(x, nd)
plt.axhline(0, alpha=0.5); plt.axvline(0, alpha=0.5)
plt.show()
```

```python
a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)
c = 20.0 * np.random.rand(5000) - 10.0

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step', label='a')
plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled', label='b')
plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step', label='c')
plt.legend()
plt.show()
```

#### 20. Matplotlib 파이 차트 그리기

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
explode = [0, 0.10, 0, 0.10]
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode)
plt.show()
```

#### Boxplot

```python
iris = sns.load_dataset('iris')
x = iris[iris.species == 'setosa'].sepal_length
y = iris[iris.species == 'setosa'].sepal_width
plt.boxplot([x, y], labels=["Sepal length", "Sepal width"])
plt.show()
```
